## Results for Notch-Delta-Jagged including Fringe effect

Equations:

\begin{equation}
\frac{dN}{dt} = N_0 H^{S+}(I) -  N ( (k_c D + kt D_{ext}) H^{S+}(I)] + (k_c J + k_t J_{ext}) H^{S-}(I)) - \gamma N
\end{equation}

\begin{equation}
\frac{dD}{dt} = D_0 H^{S-}(I) - k_c D N H^{S+}(I)- k_t D N_{ext} - \gamma D
\end{equation}

\begin{equation}
\frac{dJ}{dt} = J_0 H^{S+}(I) - k_c J N H^{S-}(I) - k_t J N_{ext} - \gamma J 
\end{equation}

\begin{equation}
\frac{dI}{dt} = k_t N (D_{ext} H^{S+}(I) + J_{ext} H^{S-}(I)) - \gamma_I I 
\end{equation}


```python
import numpy as np
import random as rand
from PyDSTool import *
from PyDSTool.Toolbox import phaseplane as pp
import PyDSTool_aux_functions as aux
%load_ext autoreload
%autoreload 2
```

```python
nmodel = 'modelNDJF'
names       = {'N': 'Notch'
              ,'D': 'Dll'
              ,'J': 'Jag'
              ,'I': 'NICD'
              }
plt.rcParams.update({'font.size': 16}) 
DSargs      = args(name=nmodel, checklevel=2)
DSargs.pars = {'g'  : 1.0e-1                                   # degradation of a protein 
              ,'gI' : 5.0e-1                                   # degradation of NICD 
              ,'kc' : 5.0e-4                                   # rate of cis interaction
              ,'kt' : 5.0e-5                                   # rate of trans interaction
              ,'p'  : 2.0e+0, 'pj' : 5.0e+0                    # hill coeficient
              ,'s0' : 2.0e+2                                   # hill factor
              ,'Nt' : 0.0e+3, 'Dt' : 0.0e+3, 'Jt' : 0.0e+3     # external proteins
              ,'N0' : 1.4e+3, 'D0' : 1.6e+3, 'J0' : 1.2e+3     # production rate
              ,'ln' : 2.0e+0, 'ld' : 0.0e+0, 'lj' : 2.0e+0     # lambda of shifted hill function
              ,'pf' : 1.0
              ,'ldf': 3.0,    'ljf': 0.3
              }
DSargs.varspecs = {'N': 'N0*HS(I,s0,p ,ln) - N*( (kc*D + kt*Dt)*HS(I,s0,pf,ldf) + (kc*J + kt*Jt)*HS(I,s0,pf,ljf) ) - g*N'
                  ,'D': 'D0*HS(I,s0,p ,ld) - D*(  kc*N*HS(I,s0,pf,ldf) + kt*Nt ) - g*D'
                  ,'J': 'J0*HS(I,s0,pj,lj) - J*(  kc*N*HS(I,s0,pf,ljf) + kt*Nt ) - g*J'
                  ,'I': 'kt*N*( Dt*HS(I,s0,pf,ldf) + Jt*HS(I,s0,pf,ljf) ) - gI*I'
                  }
DSargs.fnspecs  = [aux.functions()]
DSargs.ics      = {'N': 0 
                  ,'D': 0
                  ,'J': 0
                  ,'I': 0
                  }
DSargs.xdomain  = {'N': [0, 2.0e+4] 
                  ,'D': [0, 2.0e+4]
                  ,'J': [0, 2.0e+4]
                  ,'I': [0, 1.0e+4]
                  }
DSargs.pdomain  = {'Dt': [-1000, 5000]
                  ,'Jt': [0, 5000]
                  }
DSargs.tdomain  = [0., 500.]
DSargs.algparams= {'init_step': 0.1}
ODE = Vode_ODEsystem(DSargs)
```

```python
freepar = 'Jt'
DSargs.pdomain['Jt'] = [-15000,7000]
DSargs.pars['Jt'] = 6000
DSargs.pars['Nt'] = 500

#---------------------------------- Fig. 3A ---------------------------------------------------------------------------
ODE = Vode_ODEsystem(DSargs)
aux.plot_phasediagram(ODE, freepar, 'Dt', range(0,1501,100), xlabel='Dt', ylabel='Jt', xlim=[0,1.5], ylim=[0,6], 
                      nmodel=nmodel, names=names, silent=True)

#---------------------------------- Fig. 3B ---------------------------------------------------------------------------
DSargs.pars['J0'] = 1.4e+3
ODE = Vode_ODEsystem(DSargs)
aux.plot_phasediagram(ODE, freepar, 'Dt', range(700,1501,100), xlabel='Dt', ylabel='Jt', xlim=[0,1.5], ylim=[0,6],
                      nmodel=nmodel, names=names, silent=True)

#---------------------------------- Fig. 3C ---------------------------------------------------------------------------
DSargs.pars['J0'] = 1.2e+3
DSargs.pars['D0'] = 2.0e+3 
ODE = Vode_ODEsystem(DSargs)
aux.plot_phasediagram(ODE, freepar, 'Dt', range(0,1501,100), xlabel='Dt', ylabel='Jt', xlim=[0,1.5], ylim=[0,6],
                      nmodel=nmodel, names=names, silent=True)

#---------------------------------- Fig. 3D ---------------------------------------------------------------------------
DSargs.pars['J0'] = 1.4e+3
DSargs.pars['D0'] = 2.0e+3 
ODE = Vode_ODEsystem(DSargs)
aux.plot_phasediagram(ODE, freepar, 'Dt', range(600,1501,100), xlabel='Dt', ylabel='Jt', xlim=[0,1.5], ylim=[0,6],
                      nmodel=nmodel, names=names, silent=True)
```

```python
DSargs.pars['Jt'] = 500
DSargs.pars['Nt'] = 500
DSargs.pars['J0'] = 1.2e+3
DSargs.pars['D0'] = 1.6e+3 
ODE = Vode_ODEsystem(DSargs)

#---------------------------------- Fig. S3B ---------------------------------------------------------------------------
freepar = 'Dt'
ODE.set(pars = {freepar: 1e-3} ) 
fp = aux.eliminate_redundants(pp.find_fixedpoints(ODE, n=4, maxsearch=1e+4, eps=1e-10),6)
ODE.set(ics  = fp[0])  
PCargs = aux.PyCont_args(nmodel, freepar, 150, saveeigen=True, maxstep=1e+2, minstep=1e-2, step=5e+1)
aux.plot_continuation(ODE, PCargs, nmodel,['N','D','J','I'], freepar, names, xmin=-1, xlim=[0,2000], off_points=True,
                      fontsize=20, normal_form_coef=False, silence=True)

#---------------------------------- Fig. S3C ---------------------------------------------------------------------------
freepar = 'Jt'
ODE.set(pars = {freepar: 1e-3, 'Dt': 200} ) 
fp = aux.eliminate_redundants(pp.find_fixedpoints(ODE, n=4, maxsearch=1e+4, eps=1e-10),6)
ODE.set(ics  = fp[0])  
PCargs = aux.PyCont_args(nmodel, freepar, 150, saveeigen=True, maxstep=1e+2, minstep=1e-2, step=5e+1)
aux.plot_continuation(ODE, PCargs, nmodel,['N','D','J','I'], freepar, names, xmin=-1, xlim=[2000,5000], off_points=True, 
                      fontsize=20, normal_form_coef=False, silence=True)
```
