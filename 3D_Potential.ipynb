```python
import numpy as np
from itertools import combinations
from random import uniform
from time import clock
from mpl_toolkits.mplot3d.axes3d import Axes3D
import timeit
plt.rcParams.update({'font.size': 22}) 

# ------------------------------------hill functions------------------------------------#
def HS(X,X0,nX,l):
    return l + (1.0-l)*Hn(X,X0,nX)
def Hn(X,X0,nX):
    return 1.0/(1.0 + (X/X0)**nX)
def Hp(X,X0,nX):
    return 1.0 - Hn(X,X0,nX)
#----------------------------------------------------------------------------------------#
def func(eqs, p, k, x):
    return eval(eqs[k], p, {'N1': x[0], 'N2': x[4], 
                            'D1': x[1], 'D2': x[5], 
                            'J1': x[2], 'J2': x[6], 
                            'I1': x[3], 'I2': x[7]})
#----------------------------------------------------------------------------------------#
def euler_maruyama(eqs, keys, t, eps, dt):
    sdt = np.sqrt(dt)
    l = len(keys)
    pts = {}     
    x = np.zeros((2*l))
    for j in range(l):
        x[j  ] = np.random.uniform(vlim[keys[j]+'1'][0],vlim[keys[j]+'1'][1],1)
        x[j+l] = np.random.uniform(vlim[keys[j]+'2'][0],vlim[keys[j]+'2'][1],1)

    g_rand = np.random.normal(loc=0, scale=1, size=(int(t/dt), 2*l))
    y = np.zeros((int(t/dt), 2*l))
    for i in range(int(t/dt)):
        for j in range(l):
            if keys[j] == 'I':
                x[j]   = x[j  ] + dt*func(eqs, p, keys[j]+'1', x)
                x[j+l] = x[j+l] + dt*func(eqs, p, keys[j]+'2', x)
            else:
                x[j  ] = x[j  ] + dt*func(eqs, p, keys[j]+'1', x) + sdt*eps*g_rand[i,j  ]
                x[j+l] = x[j+l] + dt*func(eqs, p, keys[j]+'2', x) + sdt*eps*g_rand[i,j+l]
        y[i,:] = x[:]
    return y
#----------------------------------------------------------------------------------------#
def plot_potential(H, fig_name=False):
    H = np.rot90(H)
    H = np.flipud(H)
    Z = H

    V = -log(Z/float(int(t/dt)))
    V[V==inf]=NaN
    V[V>9.5]=NaN
    X,Y = meshgrid(xedges[1:]/1000, yedges[1:]/1000)
    Z = V

    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(1, 1, 1, projection='3d')
    pc = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, alpha=0.4, cmap=cm.coolwarm, linewidth=0.1)
    ax.set_zlim(5, 14)
    pc.set_clim(np.min(V[V>0]),np.max(V[V>0]))

    cset = ax.contour(X, Y, Z, zdir='z', offset=5, cmap=cm.coolwarm)
    ax.view_init(45,-135)
    ax.set_xlim3d(0, 2.5)
    ax.set_ylim3d(0, 2.5)
    ax.set_zlim3d(5, 9)
    ax.set_xticks([0,1,2])
    ax.set_yticks([0,1,2])
    ax.set_zticks([5,6,7,8,9])
    ax.set_xlabel('N1')
    ax.set_ylabel('N2')
    ax.set_zlabel('Effective Potential')
    if fig_name:
        plt.savefig('./figures/'+fig_name+'.png', format='png', dpi=200)
        plt.savefig('./figures/'+fig_name+'.pdf', format='pdf', dpi=200)
```

```python
p           = {'g'  : 1.0e-1                                   # degradation of a protein 
              ,'gI' : 5.0e-1                                   # degradation of NICD 
              ,'kc' : 5.0e-4                                   # rate of cis interaction
              ,'kt' : 2.5e-5                                   # rate of trans interaction
              ,'p'  : 2.0e+0, 'pj' : 5.0e+0                    # hill coeficient
              ,'s0' : 2.0e+2                                   # hill factor
              ,'Nt' : 0.0e+3, 'Dt' : 0.0e+3, 'Jt' : 0.0e+3     # external proteins
              ,'N0' : 1.6e+3, 'D0' : 1.7e+3, 'J0' : 1.4e+3     # production rate
              ,'ln' : 2.0e+0, 'ld' : 0.0e+0, 'lj' : 2.0e+0     # lambda of shifted hill function
              ,'pf' : 1.0
              ,'ldf': 1.5,    'ljf': 0.67
              ,'HS': HS}
eqs = {'N1': 'N0*HS(I1,s0,p ,ln) - N1*( (kc*D1 + kt*D2)*HS(I1,s0,pf,ldf) + (kc*J1 + kt*J2)*HS(I1,s0,pf,ljf) ) - g*N1'
      ,'N2': 'N0*HS(I2,s0,p ,ln) - N2*( (kc*D2 + kt*D1)*HS(I2,s0,pf,ldf) + (kc*J2 + kt*J1)*HS(I2,s0,pf,ljf) ) - g*N2'
      ,'D1': 'D0*HS(I1,s0,p ,ld) - D1*(  kc*N1*HS(I1,s0,pf,ldf) + kt*N2*HS(I2,s0,pf,ljf) ) - g*D1'
      ,'D2': 'D0*HS(I2,s0,p ,ld) - D2*(  kc*N2*HS(I2,s0,pf,ldf) + kt*N1*HS(I1,s0,pf,ljf) ) - g*D2'
      ,'J1': 'J0*HS(I1,s0,pj,lj) - J1*(  kc*N1*HS(I1,s0,pf,ljf) + kt*N2*HS(I2,s0,pf,ljf) ) - g*J1'
      ,'J2': 'J0*HS(I2,s0,pj,lj) - J2*(  kc*N2*HS(I2,s0,pf,ljf) + kt*N1*HS(I1,s0,pf,ljf) ) - g*J2'
      ,'I1': 'kt*N1*( D2*HS(I1,s0,pf,ldf) + J2*HS(I1,s0,pf,ljf) ) - gI*I1'
      ,'I2': 'kt*N2*( D1*HS(I2,s0,pf,ldf) + J1*HS(I2,s0,pf,ljf) ) - gI*I2'
      }
vlim={'N1': [0, 4.0e+3],'N2': [0, 4.0e+3] 
     ,'D1': [0, 6.0e+3],'D2': [0, 6.0e+3]
     ,'J1': [0, 5.0e+3],'J2': [0, 5.0e+3]
     ,'I1': [0, 5.0e+2],'I2': [0, 5.0e+2]
     }
```

```python
dt = 0.1
eps= 300.0
t  = 100000
keys = ['N', 'D', 'J', 'I']
i = 0 # selecting Notch proteins levels. i=1 (Delta), i=2 (Jagged), i=3 (NICD)

#------------------------ Fig. S6A ------------------------------------------
y = euler_maruyama(eqs, keys, t, eps, dt)
H, xedges, yedges = np.histogram2d(y[:,i], y[:,i+len(keys)], bins=100)
plot_potential(H)

#------------------------ Fig. S6B ------------------------------------------
p['ldf'] = 1.0
p['ljf'] = 1.0
y = euler_maruyama(eqs, keys, t, eps, dt)
H, xedges, yedges = np.histogram2d(y[:,i], y[:,i+len(keys)], bins=100)
plot_potential(H)
```
