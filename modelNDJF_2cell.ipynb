## Results for Notch-Delta-Jagged including Fringe effect

Equations:

\begin{equation}
\frac{dN}{dt} = \gamma_N(N_0 H^+(S) - N) -  N ( (k_c D + kt D_t) [1 + a_d H^+(S)] + (k_c J + k_t J_t)[1 + a_j H^+(S)])
\end{equation}

\begin{equation}
\frac{dD}{dt} = \gamma_D (D_0 H^-(S) - D) - k_c D N [1 + a_d H^+(S)]- k_t D N_t
\end{equation}

\begin{equation}
\frac{dJ}{dt} = \gamma_J (J_0 H^+(S) - J) - k_c J N [1 + a_j H^+(S)] - k_t J N_t 
\end{equation}

\begin{equation}
\frac{dS}{dt} = k_t N (D_t [1 + a_d H^+(S)] + J_t [1 + a_j H^+(S)]) - \gamma_S S 
\end{equation}


```python
import numpy as np
import random as rand
from PyDSTool import *
from PyDSTool.Toolbox import phaseplane as pp
import PyDSTool_aux_functions as aux
%load_ext autoreload
%autoreload 2

# -----------------------------------------------------------------------------------
def phase_diagram(ODE, freepar, r_var, r, xlim, ylim, nmodel, names, figname=False, silence=False):
    x  = []
    for i in r:
        ODE.set(pars = {r_var: i, freepar: 2.0e+3} )
        fp = aux.eliminate_redundants(pp.find_fixedpoints(ODE, n=2, maxsearch=1e+3, eps=1e-10),6)
        ODE.set(ics  = fp[0]) 
        PCargs = aux.PyCont_args(nmodel, freepar, 50000, saveeigen=True, maxstep=5e+1, minstep=1e-2, step=5e-0)
        p = aux.plot_continuation(ODE, PCargs, nmodel, ['N1','D1','J1','S1'], freepar, names, showcurve=False,
                                  off_points=True, fontsize=20, xlim=xlim, LimitPoints='BP', silence=silence)

        p = p[::-1]  
        for j in range(len(p)):
            x += [[i,p[j],j]]

    x = asarray(x)
    figure(figsize=(6,5), dpi=500)
    plt.xlabel(freepar, fontsize= 18)
    plt.ylabel(r_var, fontsize= 18)
    plt.xlim(xlim)
    plt.ylim(ylim)
    for i in range(int(max(x[:,2]))+1):
        plot(x[x[:,2]==i][:,1],x[x[:,2]==i][:,0], 'k')
    if figname:
        plt.savefig('./figures/'+figname+'.pdf', format='pdf', dpi=500)
```

```python
plt.rcParams.update({'font.size': 16}) 
names       = {'N1': 'Notch','D1': 'Dll','J1': 'Jag','S1': 'NICD'
              ,'N2': 'Notch','D2': 'Dll','J2': 'Jag','S2': 'NICD'
              }
nmodel = 'modelNDJF2cell'
DSargs      = args(name=nmodel, checklevel=2)
DSargs.pars = {'g'  : 1.0e-1                                   # degradation of a protein 
              ,'gS' : 5.0e-1                                   # degradation of NICD 
              ,'kc' : 5.0e-4                                   # rate of cis interaction
              ,'kt' : 2.5e-5                                   # rate of trans interaction
              ,'pn' : 2.0e+0, 'pd' : 2.0e+0, 'pj' : 5.0e+0     # hill coeficient
              ,'s0' : 2.0e+2                                   # hill factor
              ,'N0' : 1.6e+3, 'D0' : 1.7e+3, 'J0' : 1.4e+3     # production rate
              ,'Nt' : 0.0e+3, 'Dt' : 0.0e+3, 'Jt' : 0.0e+3     #
              ,'ln' : 2.0e+0, 'ld' : 0.0e+0, 'lj' : 2.0e+0     # lambda of shifted hill function
              ,'pf' : 1.0
              ,'ldf': 3.0, 'ljf' : 0.3
              }
DSargs.varspecs = {'N1': 'N0*HS(S1,s0,pn,ln) - g*N1 - N1*( (kc*D1 + kt*(D2+Dt))*HS(S1,s0,pf,ldf) + (kc*J1 + kt*(J2+Jt))*HS(S1,s0,pf,ljf))'
                  ,'N2': 'N0*HS(S2,s0,pn,ln) - g*N2 - N2*( (kc*D2 + kt*(D1+Dt))*HS(S2,s0,pf,ldf) + (kc*J2 + kt*(J1+Jt))*HS(S2,s0,pf,ljf))'
                  ,'D1': 'D0*HS(S1,s0,pd,ld) - g*D1 - D1*( kc*N1*HS(S1,s0,pf,ldf) + kt*(Nt+N2*HS(S2,s0,pf,ldf)) )'
                  ,'D2': 'D0*HS(S2,s0,pd,ld) - g*D2 - D2*( kc*N2*HS(S2,s0,pf,ldf) + kt*(Nt+N1*HS(S1,s0,pf,ldf)) )'
                  ,'J1': 'J0*HS(S1,s0,pj,lj) - g*J1 - J1*( kc*N1*HS(S1,s0,pf,ljf) + kt*(Nt+N2*HS(S2,s0,pf,ljf)) )'
                  ,'J2': 'J0*HS(S2,s0,pj,lj) - g*J2 - J2*( kc*N2*HS(S2,s0,pf,ljf) + kt*(Nt+N1*HS(S1,s0,pf,ljf)) )'
                  ,'S1': 'kt*N1*( (D2+Dt)*HS(S1,s0,pf,ldf) + (J2+Jt)*HS(S1,s0,pf,ljf) ) - gS*S1'
                  ,'S2': 'kt*N2*( (D1+Dt)*HS(S2,s0,pf,ldf) + (J1+Jt)*HS(S2,s0,pf,ljf) ) - gS*S2'
                  }
DSargs.fnspecs  = [aux.functions()]
DSargs.ics      = {'N1': 0,'D1': 0,'J1': 0,'S1': 0
                  ,'N2': 0,'D2': 0,'J2': 0,'S2': 0
                  }
DSargs.xdomain  = {'N1': [-0.0e+3, 2.0e+5],'D1': [-0.0e+3, 1.0e+5],'J1': [-0.1e+3, 2.0e+5],'S1': [-0.0e+1, 1.0e+4]
                  ,'N2': [-0.0e+3, 2.0e+5],'D2': [-0.0e+3, 1.0e+5],'J2': [-0.1e+3, 2.0e+5],'S2': [-0.0e+1, 1.0e+4]
                  }
DSargs.pdomain  = {'Jt': [-5, 10001]
                  ,'J0': [0, 2050]
                  }
DSargs.tdomain  = [0., 500.]
DSargs.algparams= {'init_step': 0.1, 'stiff': True}
ODE = Vode_ODEsystem(DSargs)
```
