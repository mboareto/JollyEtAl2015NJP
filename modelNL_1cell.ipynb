```python
import numpy as np
import random as rand
from scipy.integrate import quad
from PyDSTool import *
from PyDSTool.Toolbox import phaseplane as pp
import PyDSTool_aux_functions as aux
import copy
%load_ext autoreload
%autoreload 2

# -----------------------------------------------------------------------------------
def HS(X,X0,nX,l):
    return l + (1.0-l)*Hn(X,X0,nX)
def Hn(X,X0,nX):
    return 1.0/(1.0 + (X/X0)**nX)

# -----------------------------------------------------------------------------------
def SDE(DSargs, vlim, dt, eps,t):
    e=np.sqrt(dt)*eps
    keys = DSargs.varspecs.keys()     
    x = {}
    for j in range(len(keys)):
        x[keys[j]]    = np.zeros((int(t/dt)+1, 1))
        x[keys[j]][0] = np.random.uniform(vlim[keys[j]][0],vlim[keys[j]][1],1)

    g_rand = np.random.normal(loc=0, scale=eps, size=(int(t/dt), len(keys)))
    for i in range(int(t/dt)):
        for j in range(len(keys)):
            x[keys[j]][i+1] = x[keys[j]][i] + dt*eval(DSargs.varspecs[keys[j]], DSargs.pars, 
                                                      {'HS': HS, 'N': x['N'][i],'L': x['L'][i]})[0] + e*g_rand[i,j]
    return x

# -----------------------------------------------------------------------------------
def func(N, p):
    eqs = {'dU/dN': '-(N0*HS(kt*N*Lt/gI,s0,p,ln) - N*( kc*L + kt*Lt + g))'
           ,'L'    : '(D0*HS(kt*N*Lt/gI,s0,p ,ld) + J0*HS(kt*N*Lt/gI,s0,pj,lj))/( kc*N + kt*Nt  + g)'
          }
    return eval(eqs['dU/dN'], p, {'HS': HS, 'L': eval(eqs['L'], p, {'HS': HS, 'N': N}), 'N': N})

# -----------------------------------------------------------------------------------
def plot_bif_potential(DSargs, vlim, dt, eps, t, xlim=[0, 4000], max_p=500000):
    fp = aux.eliminate_redundants(pp.find_fixedpoints(ODE, n=4, maxsearch=1e+4, eps=1e-10),6)
    ODE.set(ics  = fp[0])  
    PCargs = aux.PyCont_args(nmodel, 'Lt', 350, saveeigen=True, maxstep=1e+2, minstep=1e-2, step=5e+1)
    aux.plot_continuation(ODE, PCargs, nmodel,['N','L'], 'Lt', names, xmin=-1, xlim=xlim, off_points=True, fontsize=20)
    x = []
    r = np.arange(0, 8000, 50)
    for i in r:
        x += [quad(func, 0, i, args=DSargs.pars)[0]]
    
    figure()
    plot(r, array(x) - min(x), 'k-', linewidth=2)
    xlabel('Notch')
    ylabel('Potential')
    ylim(0,max_p)
    #plt.savefig('./figures/Potential'+nmodel+'_Nt'+str(Nt)+'_Lt'+str(Nt)+'.pdf', format='pdf', dpi=900)
```

```python
dt = 0.1
t = 100000
eps = 22
names       = {'N': 'Notch'
              ,'L': 'Ligands'
              }

# ----------------------------------------- Figure 2A,D ------------------------------------------
nmodel = 'ND'
plt.rcParams.update({'font.size': 16}) 
DSargs      = args(name=nmodel, checklevel=2)
DSargs.pars = {'g'  : 1.0e-1                                   # degradation of a protein 
              ,'gI' : 5.0e-1                                   # degradation of NICD 
              ,'kc' : 5.0e-4                                   # rate of cis interaction
              ,'kt' : 5.0e-5                                   # rate of trans interaction
              ,'p'  : 2.0e+0, 'pj' : 5.0e+0                    # hill coeficient
              ,'s0' : 2.0e+2                                   # hill factor
              ,'Nt' : 0.5e+3, 'Lt' : 1.1e+3                    # number of external proteins
              ,'N0' : 0.5e+3, 'D0' : 1.0e+3, 'J0' : 0.0e+3     # production rate
              ,'ln' : 2.0e+0, 'ld' : 0.0e+0, 'lj' : 2.0e+0     # lambda of shifted hill function
              }
DSargs.varspecs = {'N': 'N0*HS(kt*N*Lt/gI,s0,p ,ln) - N*( kc*L + kt*Lt ) - g*N'
                  ,'L': 'D0*HS(kt*N*Lt/gI,s0,p ,ld) + J0*HS(kt*N*Lt/gI,s0,pj,lj) - L*( kc*N + kt*Nt ) - g*L'
                  }
DSargs.fnspecs  = [aux.functions()]
DSargs.ics      = {'N': 0, 'L': 0
                  }
DSargs.xdomain  = {'N': [0, 1.0e+5], 'L': [0, 1.0e+5]
                  }
DSargs.pdomain  = {'Lt': [0, 5000]
                  }
DSargs.tdomain  = [0., 100.]
DSargs.algparams= {'init_step':0.1}
ODE = Vode_ODEsystem(DSargs)

vlim = {'N': [-500,8000], 'L': [-500,8000]}
x = SDE(copy.deepcopy(DSargs), vlim, dt, eps, t)
fp = aux.eliminate_redundants(pp.find_fixedpoints(ODE, n=4, maxsearch=1e+4, eps=1e-10),6)
stab = aux.stability(fp, ODE)
aux.nullclines(['N','L'], DSargs, stab, fp, nfp=0, vlim=vlim, maxpoints=[100,100], 
               step=1e+2, minstep=1e-0, maxstep=1e+2, pcontour=x)
plot_bif_potential(DSargs, vlim, dt, eps, t)
```
