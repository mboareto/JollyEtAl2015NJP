```python
import numpy as np
from itertools import combinations
from random import uniform
from time import clock
plt.rcParams.update({'font.size': 22}) 

# ------------------------------------hill functions------------------------------------#
def HS(X,X0,nX,l):
    return l + (1.0-l)*Hn(X,X0,nX)
def Hn(X,X0,nX):
    return 1.0/(1.0 + (X/X0)**nX)
def Hp(X,X0,nX):
    return 1.0 - Hn(X,X0,nX)

#-----------------------sum proteins of the neighbors------------------------------------#
def sum_nsignal(p,pts,i,ncells,key):
    X  = np.zeros(ncells+2)
    X[ncells+1] = pts[i][0]
    X[0] = pts[i][ncells-1]
    X[1:(ncells+1)] = pts[i]
        
    if i=='N':
        I = np.zeros(ncells+2)
        I[ncells+1] = pts['I'][0]
        I[0] = pts['I'][ncells-1]
        I[1:(ncells+1)] = pts['I']
        if key=='D':
            return 0.5*(X[0:ncells]*eval('HS(I,s0,pf,ldf)',p,{'I': I[0:ncells]}) 
                        + X[2:ncells+2]*eval('HS(I,s0,pf,ldf)',p,{'I': I[2:ncells+2]}) )
        if key=='J':
            return 0.5*(X[0:ncells]*eval('HS(I,s0,pf,ljf)',p,{'I': I[0:ncells]}) 
                        + X[2:ncells+2]*eval('HS(I,s0,pf,ljf)',p,{'I': I[2:ncells+2]}) )
    else:
        return 0.5*(X[0:ncells] + X[2:ncells+2])

#-----------------------euler method for calculating trajectory--------------------------#    
def euler_traj(p,t,dt,ncells,v_lim):
    pts = {}   
    for j in vlim.keys():
        pts[j] = np.random.uniform(vlim[j][0],vlim[j][1],ncells)  # random initial condictions

    for k in range(int(t/dt)):
        pts_new = {}
        for j in vlim.keys():
            pts_new[j] = pts[j] + dt*dfdt(p,pts,j,ncells)
        pts = pts_new
    return pts

#-----------------------sum external proteins of the neighbors---------------------------#
def dfdt(p,pts,key,ncells):
    for i in pts.keys():
        p[i+'t'] = sum_nsignal(p,pts,i,ncells,key) 
    return eval(eqs[key[0]], p, pts)
```

```python
p = {'g'  : 1.0e-1                                   # degradation of a protein 
    ,'gI' : 5.0e-1                                   # degradation of NICD 
    ,'kc' : 5.0e-4                                   # rate of cis interaction
    ,'kt' : 2.5e-5                                   # rate of trans interaction
    ,'p'  : 2.0e+0, 'pj' : 5.0e+0                    # hill coeficient
    ,'s0' : 2.0e+2                                   # hill factor
    ,'N0' : 1.6e+3, 'D0' : 1.8e+3, 'J0' : 1.2e+3     # production rate
    ,'ln' : 2.0e+0, 'ld' : 0.0e+0, 'lj' : 2.0e+0     # lambda of shifted hill function
    ,'pf' : 1.0
    ,'ldf': 3.0,    'ljf': 0.3
    ,'HS': HS, 'Hp': Hp, 'Hn': Hn
    }
vlim={'N': [0,600], 'D': [0,600], 'J': [0,600], 'I': [0,60]}
eqs = {'N': 'N0*HS(I,s0,p ,ln) - N*( (kc*D + kt*Dt)*HS(I,s0,pf,ldf) + (kc*J + kt*Jt)*HS(I,s0,pf,ljf) ) - g*N'
      ,'D': 'D0*HS(I,s0,p ,ld) - D*(  kc*N*HS(I,s0,pf,ldf) + kt*Nt ) - g*D'
      ,'J': 'J0*HS(I,s0,pj,lj) - J*(  kc*N*HS(I,s0,pf,ljf) + kt*Nt ) - g*J'
      ,'I': 'kt*N*( Dt*HS(I,s0,pf,ldf) + Jt*HS(I,s0,pf,ljf) ) - gI*I'
      }
```

```python
ncells = 100
ncomb = 1
dt = 0.1
t  = 100

X = {}
for i in vlim.keys():
    X[i]=[]

r = np.arange(1,3.1,0.5)
for k in r:
    p['ldf'] = k
    p['ljf'] = 1.0/k
    for i in range(ncomb):
        pts = euler_traj(p,t,dt,ncells,vlim)
        for j in vlim.keys():
            X[j] += [list(pts[j])]

figure(figsize=(11,3))  
x = X['I']
x = asarray(x)
x[x<100] = 0
x[x>300] = 1
x[x>100] = 0.45
pcolor(x,edgecolors='k', linewidths=3, vmin=-0.11, vmax=1.015)
#plt.savefig('./figures/1D_patterningNDJF_J0.pdf', format='pdf', dpi=900)
```
